```python
import graphviz
```

```python
def make_dotcell( thegraph, format="svg" ):
    cell_content = "%%dot " + "-f %s\n%s" % (format, thegraph.source)
    return cell_content
```

```python
tribe = ["Asterix", "Obelix", "Idefix", "Miraculix", "Troubadix"]
places = ["Gaul", "Rome"]

Asterix_and_Obelix = graphviz.Digraph()

Asterix_and_Obelix.graph_attr = {"bgcolor": "black"}
Asterix_and_Obelix.node_attr = {'fontcolor': "white" }
Asterix_and_Obelix.edge_attr = {'fontcolor': "white", 'color': "yellow"}

for ix in tribe:
    Asterix_and_Obelix.node(ix)

for place in places:
    Asterix_and_Obelix.node(place, shape="house", color="red", fontcolor="white")

Asterix_and_Obelix.edge( "Asterix", "Obelix", label="is friend", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Idefix", "Asterix", label="is dog of", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Idefix", "Obelix", label="is dog of", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Miraculix", "Asterix", label="lives in the same village", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Troubadix", "Asterix", label="lives in the same village", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Miraculix", "Obelix", label="lives in the same village", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Troubadix", "Obelix", label="lives in the same village", color="green", fontcolor="green" )
Asterix_and_Obelix.edge( "Asterix", "Rome", label="goes to" )
Asterix_and_Obelix.edge( "Obelix", "Rome", label="goes to" )
Asterix_and_Obelix.edge( "Idefix", "Gaul", label="stays in" )

%recall make_dotcell( Asterix_and_Obelix )
```

```python
%%dot -f svg
digraph {
	graph [bgcolor=black]
	node [fontcolor=white]
	edge [color=yellow fontcolor=white]
		Asterix
		Obelix
		Idefix
		Miraculix
		Troubadix
		Gaul [color=red fontcolor=white shape=house]
		Rome [color=red fontcolor=white shape=house]
			Asterix -> Obelix [label="is friend" color=green fontcolor=green]
			Idefix -> Asterix [label="is dog of" color=green fontcolor=green]
			Idefix -> Obelix [label="is dog of" color=green fontcolor=green]
			Miraculix -> Asterix [label="lives in the same village" color=green fontcolor=green]
			Troubadix -> Asterix [label="lives in the same village" color=green fontcolor=green]
			Miraculix -> Obelix [label="lives in the same village" color=green fontcolor=green]
			Troubadix -> Obelix [label="lives in the same village" color=green fontcolor=green]
			Asterix -> Rome [label="goes to"]
			Obelix -> Rome [label="goes to"]
			Idefix -> Gaul [label="stays in"]
}
```
